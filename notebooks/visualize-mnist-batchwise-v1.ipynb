```python
import json

import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("../experiment-data/1/info.json", "r") as f:
    data = json.load(f)
with open("../experiment-data/1/config.json", "r") as f:
    config = json.load(f)
    scales = config['scales']
```

```python
peak_train_accuracy = []
peak_test_accuracy = []
for scale in scales:
    results = data[f'json://{scale}']
    peak_train_accuracy.append(max(results['train_accuracies']))
    peak_test_accuracy.append(max(results['test_accuracies']))

plt.plot(scales, peak_train_accuracy, label="train")
plt.plot(scales, peak_test_accuracy, label="test")
plt.ylabel("max accuracy", fontsize=14)
plt.xlabel("noise scale", fontsize=14)
plt.title("Accuracy phase transition with batch-wise noisy parameters")
plt.legend()
```

```python
peak_train_accuracy = []
peak_test_accuracy = []
for scale in scales:
    results = data[f'json://{scale}']
    peak_train_accuracy.append(max(results['train_accuracies']))
    peak_test_accuracy.append(max(results['test_accuracies']))

plt.plot(scales, peak_train_accuracy, label="train")
plt.plot(scales, peak_test_accuracy, label="test")
plt.ylabel("max accuracy", fontsize=14)
plt.xlabel("noise scale", fontsize=14)
plt.xscale('log')
plt.xlim(0.12, 0.25)
plt.title("Accuracy phase transition with batch-wise noisy parameters")
plt.legend()
```

```python
peak_train_loss = []
peak_test_loss = []
for scale in scales:
    results = data[f'json://{scale}']
    peak_train_loss.append(min(results['train_losses']))
    peak_test_loss.append(min(results['test_losses']))

plt.plot(scales, peak_train_loss, label="train")
plt.plot(scales, peak_test_loss, label="test")
plt.ylabel("min loss", fontsize=14)
plt.xlabel("noise scale", fontsize=14)
plt.title("loss phase transition with batch-wise noisy parameters")
plt.legend()
```
